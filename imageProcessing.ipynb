```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('C:\\Users\\admin\\Desktop\\image.jpg')
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show(block=True)
```

```python
print("Original Image Dimensions:")
print(f"Height: {img.shape[0]}, Width: {img.shape[1]}, Channels: {img.shape[2]}")
```

```python
resized_img=cv2.resize(img, (400, 256))
```

```python
plt.imshow(cv2.cvtColor(resized_img,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
horizontal_flip = cv2.flip(img, 1)
```

```python
plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(horizontal_flip,cv2.COLOR_BGR2RGB))
plt.title('Horizontal Flip')
```

```python
vertical_flip = cv2.flip(img, 0)
```

```python
plt.subplot(2,2,3)
plt.imshow(cv2.cvtColor(vertical_flip,cv2.COLOR_BGR2RGB))
plt.title('Vertical Flip')
```
