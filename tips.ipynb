```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
file_path = r'C:\Users\Admin\Desktop\tips.csv'
```

```python
tips_df = pd.read_csv(file_path)
```

```python
smoking_count = tips_df[tips_df['Smoking'] == 'Yes'].shape[0]
non_smoking_count = tips_df[tips_df['Smoking'] == 'No'].shape[0]
```

```python
smoking_percentage = (smoking_count / tips_df.shape[0]) * 100
non_smoking_percentage = (non_smoking_count / tips_df.shape[0]) * 100
```

```python
print(f"Smoking area: {smoking_percentage:.2f}%")
print(f"Non-smoking area: {non_smoking_percentage:.2f}%")
```

```python
plt.figure(figsize=(8, 8))
plt.pie([smoking_count, non_smoking_count], labels=['Smoking', 'Non-smoking'], autopct='%1.1f%%')
plt.title('Percentage of People in Smoking and Non-smoking Areas')
plt.show()
```
